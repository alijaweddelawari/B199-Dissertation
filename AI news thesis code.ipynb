```python


# Loading the dataset
df = pd.read_excel("newssheet.xlsx")

# Droping the first blank row from the dataset
df = df.drop(index=0)

# Rename columns correctly
df.columns = ["ID", "Source", "Date", "Headline", "URL", "Summary", "Sentiment"]

# Reset index
df = df.reset_index(drop=True)

# Drop rows where both Headline AND Summary are empty
df = df.dropna(subset=["Headline", "Summary"], how="all")

# Remove duplicates based on headline + summary
df = df.drop_duplicates(subset=["Headline", "Summary"])

df.head(10)
```

```python
# Converting Date column to datetime and remove time
df["Date"] = pd.to_datetime(df["Date"], errors="coerce").dt.date

df.head(10)
```

```python
# Removeing completely empty rows
df = df.dropna(how="all")

# Removeing rows where Summary is missing
df = df.dropna(subset=["Summary"])
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    if isinstance(text, str):
        text = text.lower()
        text = re.sub(r"http\S+", "", text)  # remove URLs
        text = re.sub(r"[^a-zA-Z\s]", "", text)  # remove punctuation/numbers
        text = text.strip()
        words = text.split()
        words = [w for w in words if w not in stop_words]
        words = [lemmatizer.lemmatize(w) for w in words]
        return " ".join(words)
    return text

df["clean_summary"] = df["Summary"].apply(clean_text)

df.head()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

df["sentiment_score"] = df["clean_summary"].apply(lambda x: sia.polarity_scores(x)["compound"])

def get_sentiment_label(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df["sentiment_label"] = df["sentiment_score"].apply(get_sentiment_label)

df[["Summary", "clean_summary", "sentiment_score", "sentiment_label"]].head()
```

```python
# Installing gensim here
!pip install gensim

from gensim import corpora, models
import nltk
nltk.download("stopwords")

from nltk.corpus import stopwords
stop_words = stopwords.words("english")

# Preparing data for LDA
texts = df["clean_summary"].apply(lambda x: x.split()).tolist()

# Creating dictionary and corpus
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Training LDA model with 3 topics
lda_model = models.LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=3,
    passes=15,
    random_state=42
)

# Showing topics
topics = lda_model.print_topics(num_words=8)
for idx, topic in topics:
    print(f"Topic {idx}: {topic}")
```

```python
# Assigning dominant topic to each document
def get_dominant_topic(lda_model, corpus):
    dominant_topics = []
    for doc in corpus:
        topic_probs = lda_model.get_document_topics(doc)
        dominant_topic = max(topic_probs, key=lambda x: x[1])[0]
        dominant_topics.append(dominant_topic)
    return dominant_topics

df["dominant_topic"] = get_dominant_topic(lda_model, corpus)
df[["Summary", "dominant_topic"]].head()
```

```python
# Afghan AI News – Interactive Result Exploration Module
# Purpose: Dataset inspection, result validation, and analytical exploration

import pandas as pd

# Helper Functions

def inspect_article(df):
    try:
        article_id = int(input("Enter Article ID: "))
        row = df[df["ID"] == article_id]

        if row.empty:
            print("\n❗ No article found with that ID.\n")
        else:
            print("\n===== Article Inspection =====")
            print(f"ID: {row['ID'].values[0]}")
            print(f"Source: {row['Source'].values[0]}")
            print(f"Date: {row['Date'].values[0]}")
            print(f"Headline: {row['Headline'].values[0]}")
            print(f"URL: {row['URL'].values[0]}")
            print(f"Summary: {row['Summary'].values[0]}")
            print(f"Assigned Topic: {row['topic_label'].values[0]}")
            print(f"Assigned Sentiment: {row['sentiment_label'].values[0]}")
            print("==============================\n")

    except ValueError:
        print("\n❗ Please enter a valid numeric ID.\n")


def filter_by_topic(df):
    print("\nAvailable Topics:")
    for t in df["topic_label"].unique():
        print("-", t)

    selected = input("\nEnter topic name exactly as shown: ")
    results = df[df["topic_label"] == selected]

    if results.empty:
        print("\n❗ No articles found for that topic.\n")
    else:
        print(f"\n===== Articles under '{selected}' ({len(results)}) =====")
        print(results[["ID", "Headline", "sentiment_label"]])
        print("===============================================\n")


def filter_by_sentiment(df):
    sentiments = ["Positive", "Neutral", "Negative"]
    print("\nAvailable Sentiment Labels:", sentiments)

    selected = input("\nEnter sentiment label: ").capitalize()

    if selected not in sentiments:
        print("\n❗ Invalid sentiment label.\n")
        return

    results = df[df["sentiment_label"] == selected]

    if results.empty:
        print("\n❗ No articles match this sentiment.\n")
    else:
        print(f"\n===== {selected} Articles ({len(results)}) =====")
        print(results[["ID", "Headline", "topic_label"]])
        print("=====================================\n")


def filter_by_topic_and_sentiment(df):
    topic = input("Enter topic: ")
    sentiment = input("Enter sentiment (Positive/Neutral/Negative): ").capitalize()

    results = df[
        (df["topic_label"] == topic) &
        (df["sentiment_label"] == sentiment)
    ]

    if results.empty:
        print("\n❗ No matching articles found.\n")
    else:
        print(f"\n===== '{topic}' + '{sentiment}' ({len(results)}) =====")
        print(results[["ID", "Headline"]])
        print("=====================================\n")


def show_statistics(df):
    print("\n===== Dataset-Level Statistics =====\n")

    print("Articles per Topic:")
    print(df["topic_label"].value_counts(), "\n")

    print("Articles per Sentiment:")
    print(df["sentiment_label"].value_counts(), "\n")

    print("Topic × Sentiment Matrix:")
    print(df.groupby(["topic_label", "sentiment_label"]).size().unstack(fill_value=0))

    print("\n===================================\n")


# Main Exploration Loop

def run_explorer(df):
    while True:
        print("""
=== Afghan AI News – Result Inspection Interface ===

1. Inspect article by ID
2. Filter articles by Topic
3. Filter articles by Sentiment
4. Filter by Topic + Sentiment
5. Show Summary Statistics
0. Exit
        """)

        choice = input("Select an option: ")

        if choice == "1":
            inspect_article(df)
        elif choice == "2":
            filter_by_topic(df)
        elif choice == "3":
            filter_by_sentiment(df)
        elif choice == "4":
            filter_by_topic_and_sentiment(df)
        elif choice == "5":
            show_statistics(df)
        elif choice == "0":
            print("\nExiting Result Explorer.\n")
            break
        else:
            print("\n❗ Invalid option. Please try again.\n")


# Run the exploration module
run_explorer(df)
```

```python

# Text analysis helper

def analyze_text(text):
    cleaned = clean_text(text)
    score = sia.polarity_scores(cleaned)
    
    # choose label
    if score["compound"] >= 0.05:
        sentiment = "Positive"
    elif score["compound"] <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    # prepare LDA topic prediction
    bow = dictionary.doc2bow(cleaned.split())
    topic_dist = lda_model.get_document_topics(bow)
    topic_dist = sorted(topic_dist, key=lambda x: x[1], reverse=True)

    if topic_dist:
        dominant_topic = topic_dist[0][0]
        keywords = [word for word, _ in lda_model.show_topic(dominant_topic, topn=5)]
    else:
        dominant_topic = None
        keywords = []

    print("\n===== Text Analysis Result =====")
    print("Cleaned Text:", cleaned)
    print("Sentiment:", sentiment, f"(score={score['compound']})")
    print("Topic:", dominant_topic, " | Keywords:", keywords)
    print("================================\n")
```

```python
# 3. Map 1–5 sentiment scale → Negative / Neutral / Positive
def map_sentiment(x):
    if x in [1, 2]:
        return "Negative"
    elif x == 3:
        return "Neutral"
    elif x in [4, 5]:
        return "Positive"

df_sample["true_sentiment"] = df_sample["Sentiment"].apply(map_sentiment)

# 4. Cleaning the headlines
df_sample["clean_headline"] = df_sample["Headlines"].apply(clean_text)

# 5. Appling VADER sentiment analysis via NLTK
sia = SentimentIntensityAnalyzer()

def vader_predict(text):
    scores = sia.polarity_scores(text)
    c = scores["compound"]
    if c >= 0.05:
        return "Positive"
    elif c <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df_sample["predicted_sentiment"] = df_sample["clean_headline"].apply(vader_predict)

# 6. Evaluating performance
accuracy = accuracy_score(df_sample["true_sentiment"], df_sample["predicted_sentiment"])
print("\nAccuracy:", round(accuracy * 100, 2), "%")

print("\nClassification Report:")
print(classification_report(df_sample["true_sentiment"], df_sample["predicted_sentiment"]))
```
